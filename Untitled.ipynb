```python
import re

def peak_list_str_parser(res_str):    
    res_letter=res_str[0]
    integers=[int(s) for s in re.findall(r'-?\d+\.?\d*', res_str)]
    if(res_letter=='I'):
        name='ILE'+' '+str(integers[0])+' A'
    elif(res_letter=='A'):
        name='ALA'+' '+str(integers[0])+' A'
    elif(res_letter=='L'):
        name='LEU'+' '+str(integers[0])+' A'
    elif(res_letter=='V'):
        name='VAL'+' '+str(integers[0])+' A'
    return name

res1,res2='I123','V123'
res1,res2=peak_list_str_parser(res1),peak_list_str_parser(res2)
print res1,res2
    
```
